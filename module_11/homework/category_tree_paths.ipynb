### ЗАДАЧА: Дерево категорий товаров

Есть список путей категорий, например `Электроника/Ноутбуки/Игровые`.
Нужно собрать из этих путей одно дерево каталога.

Общие части пути должны переиспользоваться, а не создаваться заново.

НЕОБХОДИМО РЕАЛИЗОВАТЬ:

1. Класс `TreeNode`, который хранит:
   - имя узла `name`,
   - словарь дочерних узлов `children`.

2. Функцию `add_path(root, path)`, которая добавляет путь в дерево.

3. Функцию `count_nodes(root)` — общее количество узлов.

4. Функцию `count_leaves(root)` — количество листьев.

5. Функцию `get_all_leaf_paths(root)`, которая возвращает все полные пути до листьев.

6. После построения дерева вывести:
   - количество узлов,
   - количество листьев,
   - список всех конечных путей.


```python
from dataclasses import dataclass, field

paths = [
    "Электроника/Ноутбуки/Игровые",
    "Электроника/Ноутбуки/Ультрабуки",
    "Электроника/Телефоны/Android",
    "Электроника/Телефоны/iPhone",
    "Дом/Кухня/Посуда",
    "Дом/Кухня/Техника",
    "Дом/Мебель/Стулья",
]

@dataclass
class TreeNode:
    name: str
    children: dict[str, "TreeNode"] = field(default_factory=dict)

def add_path(root: TreeNode, path: str) -> None:
    parts = path.split('/')
    current = root
    for part in parts:
        if part not in current.children:
            current.children[part] = TreeNode(part)
        current = current.children[part]

def count_nodes(root: TreeNode) -> int:
    count = 1  # Считаем текущий узел
    for child in root.children.values():
        count += count_nodes(child)
    return count

def count_leaves(root: TreeNode) -> int:
    if not root.children:  # Если у узла нет детей — это лист
        return 1
    total_leaves = 0
    for child in root.children.values():
        total_leaves += count_leaves(child)
    return total_leaves

def get_all_leaf_paths(root: TreeNode, prefix: str = "") -> list[str]:
    paths = []
    # Формируем текущий путь
    current_path = f"{prefix}/{root.name}" if prefix else root.name
    
    if not root.children:  # Если узел — лист, добавляем путь в результат
        paths.append(current_path)
    else:
        # Рекурсивно обходим всех детей
        for child in root.children.values():
            paths.extend(get_all_leaf_paths(child, current_path))
    return paths

root = TreeNode("Каталог")

for path in paths:
    add_path(root, path)

print("Всего узлов:", count_nodes(root))
print("Всего листьев:", count_leaves(root))
print("Конечные пути:")
for path in get_all_leaf_paths(root):
    print(path)
```

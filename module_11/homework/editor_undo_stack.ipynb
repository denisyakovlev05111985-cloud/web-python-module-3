### ЗАДАЧА: Undo в текстовом редакторе через стек

Нужно реализовать простой редактор с командами:
- `append:текст`,
- `undo`,
- `show`.

Для отката используйте **стек** состояний.
Если стек пустой, команда `undo` не должна менять текст.

НЕОБХОДИМО РЕАЛИЗОВАТЬ:

1. Использовать список как стек состояний.

2. Перед каждым `append` сохранять предыдущее состояние текста в стек.

3. При `undo` доставать последнее состояние из стека.

4. Собрать все результаты команд `show` в список `snapshots`.

5. Посчитать:
   - финальный текст,
   - количество успешных `undo`,
   - максимальную глубину стека за всё время.

6. Вывести `snapshots`, финальный текст и статистику.


```python
commands = [
    "append:Привет",
    "append:, мир",
    "show",
    "append:!",
    "show",
    "undo",
    "show",
    "append:!!!",
    "undo",
    "undo",
    "undo",
    "show",
]

text = ""
state_stack = []
snapshots = []
undo_count = 0
max_stack_size = 0

for command in commands:
    if command.startswith("append:"):
        value = command.split(":", 1)[1]
        # Сохраняем текущее состояние текста в стек
        state_stack.append(text)
        # Обновляем максимальную глубину стека, если текущий размер больше
        max_stack_size = max(max_stack_size, len(state_stack))
        # Добавляем value к text
        text += value

    elif command == "undo":
        # Если стек не пуст, достаём последнее состояние
        if state_stack:
            text = state_stack.pop()
            # Увеличиваем счётчик успешных undo
            undo_count += 1

    elif command == "show":
        # Добавляем текущее состояние текста в snapshots
        snapshots.append(text)

print("Снимки:", snapshots)
print("Финальный текст:", text)
print("Количество undo:", undo_count)
print("Максимальная глубина стека:", max_stack_size)
```
